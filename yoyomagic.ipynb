```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.ndimage import gaussian_filter
def blur(img, dpi):
    return gaussian_filter(img, (3,3,0)), -0, 0
```

```python
from scipy.ndimage import gaussian_filter
def Blur(sigma):
    def blur(img, dpi):
        return gaussian_filter(img, (sigma,sigma,0)), -0, 0
    return blur
```

```python
def plotx(*a, linewidth=2, alpha=1.0, **k):
    plt.plot(*a, **k, linewidth=12*linewidth, alpha=alpha*0.3, agg_filter=Blur(5))
    plt.plot(*a, **k, linewidth=5*linewidth, alpha=alpha*0.6, agg_filter=Blur(3))
    plt.plot(*a, **k, linewidth=linewidth, alpha=alpha)
```

```python
import matplotlib.font_manager
```

```python
plt.figure(figsize=(5,5))


ax = plt.gca()
ax.set_facecolor('#000000')

t = np.linspace(0, 2*np.pi, 101)



plt.plot(.9*np.cos(t), .9*np.sin(t), '#E8CCFF',linewidth=7, agg_filter=blur)


plt.plot(1.2*np.cos(t), 1.2*np.sin(t), '#FFFF77',linewidth=5,agg_filter=blur)


plt.plot(1.7*np.cos(t), 1.7*np.sin(t),'--',linewidth=5,agg_filter=blur)

plt.plot(3*np.cos(t), 3*np.sin(t),'#FFFFFF',linewidth=1)

plt.plot(2.2*np.cos(t), 2.2*np.sin(t),'#FFFFFF',linewidth=1)


for i in range(6):
    t0 = 2*np.pi/6*i
    x0,y0 = 2*np.cos(t0), 2*np.sin(t0)
    plt.plot(.8*np.cos(t)+x0, .8*np.sin(t)+y0, '#E8CCFF',linewidth=2,agg_filter=blur)


    

TEXT = 'MAPLESTORY'
font0 = matplotlib.font_manager.FontProperties(family='serif', 
            size=26, style='italic', weight='heavy')
font = matplotlib.font_manager.FontProperties(family='serif', 
            size=20, style='italic', weight='light')
for i in range(len(TEXT)):
    s = 1*np.pi*2*i/len(TEXT)
    x0,y0 = 1.4*np.cos(s), 1.4*np.sin(s)
    ax.text(x0, y0, TEXT[i], 
            color='#FF0000', va='center', ha='center', 
            rotation=s*180/np.pi-90, fontproperties=font)
    ax.text(x0, y0, TEXT[i], 
            color='#FFFFFF', va='center', ha='center', 
            rotation=s*180/np.pi-90, fontproperties=font0,agg_filter=Blur(3))
    plt.plot(.24*np.cos(t)+x0, .24*np.sin(t)+y0, '#ffffff',linewidth=2,agg_filter=blur)

color = (.8,1,1)
t = np.linspace(0, 2*np.pi, 5)
plotx(1.2*np.cos(t), 1.2*np.sin(t), color=color, alpha=0.8)
plotx(1.2*np.cos(t+np.pi/4), 1.2*np.sin(t+np.pi/4), color='w', alpha=0.6)
    
color = (.8,1,1)
t = np.linspace(0, 2*np.pi, 5)
plotx(2*np.cos(t), 2*np.sin(t), color=color, alpha=0.8)
plotx(2*np.cos(t+np.pi/4), 2*np.sin(t+np.pi/4), color='w', alpha=0.6)

t = np.linspace(0, 3*np.pi, 101)
for d in range(5):
    shift = np.random.uniform(low=0, high=2*np.pi)
    noise = np.random.uniform(low=-0.05, high=0.05, size=t.shape)
    freq = np.random.uniform(low=0.5, high=2)
    r = np.random.uniform(low=0.5, high=2, size=t.shape)
    color = tuple(np.random.uniform(0.8,1, size=3))
    z = 3+.05*r*np.sin(freq*t+shift+noise)+noise
    plt.plot(z*np.cos(t), z*np.sin(t), c=color, linewidth=12,
             alpha=0.1, agg_filter=Blur(7))
    plt.plot(z*np.cos(t), z*np.sin(t), c=color, linewidth=7, alpha=0.3, 
             agg_filter=Blur(5))
    plt.plot(z*np.cos(t), z*np.sin(t), c=color, linewidth=2, alpha=.5, 
             agg_filter=Blur(1))

a = np.linspace(0,10*np.pi,12)
plt.plot(.9*np.cos(a+np.pi/10), .9*np.sin(a+np.pi/10), '#00FFFF',linewidth=2,agg_filter=blur)
plt.savefig('yomagic.png')
```
